```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
```

```python
loansData = pd.read_csv('https://spark-public.s3.amazonaws.com/dataanalysis/loansData.csv') # creates dataframe
```

```python
# remove nulls
loansData.dropna(inplace=True)
```

```python
# create the boxplot
loansData.boxplot(column='Amount.Requested')

plt.show()
```

```python
# do the histogram
loansData.hist(column='Amount.Requested')
plt.show()
```

```python
# Generate the QQ
plt.figure()
graph = stats.probplot(loansData['Amount.Requested'], dist="norm", plot=plt)
plt.show()
```
