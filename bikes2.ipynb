```python
import requests
```

```python
r = requests.get('http://www.citibikenyc.com/stations/json')
```

```python

intlist = ['availableDocks', 'totalDocks', 'statusKey', 'availableBikes', 'id']
```

```python
 for index, station in enumerate(r.json()['stationBeanList']):
     for count, item in enumerate(intlist):
         if r.json()['stationBeanList'][index][intlist[count]] == 0:
             print(r.json()['stationBeanList'][index]['id'])
```
